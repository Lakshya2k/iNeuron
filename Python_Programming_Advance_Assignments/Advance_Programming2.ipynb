## Q1

```python
def pentagonal(num):
    return 1 + 5 * (num - 1) * num // 2
```

```python
pentagonal(8)
```

## Q2

```python
def encrypt(phrase):
    vowels = {"a": "0", "e": "1", "i": "2", "o": "2", "u": "3"}
    for letter in phrase:
        if letter in vowels.keys():
            phrase = phrase.replace(letter, vowels[letter])
    return phrase[::-1] + "aca"
```

```python
encrypt("banana")
```

## Q3

```python
import datetime

def has_friday_13(mon, yr):
    return datetime.date(yr, mon, 13).weekday()==4
```

```python
has_friday_13(10, 2017)
```

## Q4

```python
import re

def count_bad_cookies(l):
    return len(re.findall("bad cookie", ",".join(l)))
```

```python
count_bad_cookies(["bad cookie", "good cookie", "bad cookie", "good cookie", "good cookie"])
```

## Q5

```python
from collections import Counter

def pluralize(l):
    return {k+"s" if v>1 else k for k,v in Counter(l).items()}
```

```python
pluralize(["cow", "pig", "cow", "cow"])
```

```python
pluralize(["chair", "pencil", "arm"])
```
