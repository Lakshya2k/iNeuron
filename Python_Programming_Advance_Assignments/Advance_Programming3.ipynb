## Q1

```python
def arithmetic_operation(expr):
    arr = expr.split(" ")
    num_1 = int(arr[0])
    opr = arr[1]
    num_2 = int(arr[2])
    if opr == '+' :
        return num_1 + num_2
    elif opr == '-' :
        return num_1 - num_2
    elif opr == '*' :
        return num_1 * num_2
    else:
        if num_2 == 0 :
            return -1
        else :
            return num_1 // num_2
```

```python
arithmetic_operation('12 + 12')
```

```python
arithmetic_operation('12 - 12')
```

```python
arithmetic_operation('12 * 12')
```

```python
arithmetic_operation('12 // 0')
```

## Q2

```python
def perimeter(pts): 
    return ((pts[0][0]-pts[1][0])**2 + (pts[0][1]-pts[1][1])**2)**0.5 + ((pts[1][0]-pts[2][0])**2 + (pts[1][1]-pts[2][1])**2)**0.5 + ((pts[2][0]-pts[0][0])**2 + (pts[2][1]-pts[0][1])**2)**0.5
```

```python
perimeter([[15,7],[5,22],[11,1]])
```

## Q3

```python
def tallest_skyscraper(arr):
    tallest = arr[0][0] + arr[1][0] + arr[2][0] + arr[3][0]
    for i in range(1,4):
        temp = arr[0][i] + arr[1][i] + arr[2][i] + arr[3][i]
        if temp > tallest:
            tallest = temp
    return tallest
```

```python
tallest_skyscraper([
    [0,1,0,0],
    [0,1,0,0],
    [0,1,1,0],
    [1,1,1,1]
])
```

## Q4

```python
def bonus(days):
    if days<=32:
        return 0
    elif days>32 and days<=40:
        return (days-32)*325
    elif days>40 and days<=48:
        return (days-32)*325 + (days-8)*550
    else:
        return 8*325 + 8*550 + (days-48)*600
```

```python
bonus(37)
```

```python
bonus(50)
```

## Q5

```python
def is_disarium(num):
    sum = 0
    for i,e in enumerate(str(num)):
        sum += int(e)**(i+1)
    return sum == num
```

```python
is_disarium(135)
```
