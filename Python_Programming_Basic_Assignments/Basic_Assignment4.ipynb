## Q1

```python
def factorial(num):
    return 1 if (num==0 or num==1) else num*factorial(num-1)
```

```python
factorial(5)
```

```python
factorial(9)
```

## Q3

```python
def fibonacci(end):
    n1,n2,count=0,1,0
    if end<=0:
        print("invalid input")
    elif end==1:
        print(n1)
    else:
        while count<end:
            print(n1, end=" ")
            nth = n1 + n2
            n1,n2 = n2,nth
            count += 1
            
```

```python
fibonacci(10)
```

## Q4

```python
def check_armstrong(num):
    n = len(str(num))
    sum = 0
    temp = num
    while temp>0:
        digit = temp%10
        sum += digit**n
        temp = temp//10
    if num==sum:
        print("armstrong")
    else:
        print("non-armstrong")
```

```python
check_armstrong(153)
```

```python
check_armstrong(1234)
```

## Q5

```python
def print_armstrong(lo, hi):
    for num in range(lo, hi+1):
        n = len(str(num))
        sum = 0
        temp = num
        while temp>0:
            digit = temp%10
            sum += digit**n
            temp = temp//10
        if num==sum:
            print(num, end=" ")
```

```python
print_armstrong(100,1000)
```
