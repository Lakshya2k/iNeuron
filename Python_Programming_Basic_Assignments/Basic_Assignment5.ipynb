## Q1

```python
def find_lcm(x, y):
    greater = x if x>y else y
    while(True):
        if greater%x==0 and greater%y==0:
            lcm = greater
            break
        greater += 1
    return lcm
```

```python
find_lcm(18, 90)
```

```python
find_lcm(19, 361)
```

## Q2

```python
def find_hcf(x, y):
    smaller = x if x<y else y
    for i in range(1, smaller+1):
        if x%i==0 and y%i ==0:
            hcf = i
    return hcf
```

```python
find_hcf(7, 343)
```

```python
find_hcf(20, 184)
```

## Q5

```python
def calc(x, y, op):
    if op=="+":
        return x+y
    if op=="-":
        return x-y
    if op=="*":
        return x*y
    if op=="/":
        return x/y
```

```python
calc(7,9,)
```
