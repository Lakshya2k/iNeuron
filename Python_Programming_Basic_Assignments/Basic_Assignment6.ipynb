## Q1

```python
def show_fib(n):
    def calc_fib(num):
        if num<=1:
            return num
        else:
            return calc_fib(num-1)+calc_fib(num-2)
    if n<=0:
        print("invald input")
    else:
        for i in range(n):
            print(calc_fib(i), end=" ")
```

```python
show_fib(10)
```
