## Q1

```python
def evenly_divisible(a, b, c):
    total = 0
    for i in range(a, b+1):
        if i%c == 0:
            total = total + i
    return total
```

```python
evenly_divisible(1,10,3)
```

## Q3

```python
def replace_vowels(st, c):
    vowels = "aeiouAEIOU"
    for vowel in vowels:
        st = st.replace(vowel, c)
    return st
```

```python
replace_vowels("minnie mouse", "?")
```

## Q4

```python
def factorial(num):
    if num == (0 or 1):
        return 1
    else:
        return num*factorial(num-1)
```

```python
factorial(5)
```

## Q5

```python
def hamming_distance(a, b):
    d = 0
    for x,y in zip(a, b):
        if x!=y:
            d = d + 1
    return d 
```

```python
hamming_distance("strong", "strung")
```
