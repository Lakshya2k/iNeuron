## Q1

```python
def string_gtk(k, s):
    text = s.split(" ")
    s1 = []
    for x in text:
        if len(x)>k:
            s1 = s1 + [x]
    return s1
```

```python
string_gtk(4, "hello nice meeting")
```

```python
def split_string(s):
    return s.split(" ")
def join_string(ls):
    return " ".join(ls)
```

```python
split_string("This is an apple")
```

```python
join_string(["This", "is", "an", "apple"])
```

## Q4

```python
def check_binary(s):
    s1 = set(s)
    s2 = {"0","1"}
    if s1==s2 or s1=={"0"} or s1=={"1"}:
        return True
    else:
        return False
```

```python
check_binary("010101011111100000")
```

```python
check_binary("Hello")
```

## Q5

```python
def uncommon_words(s1, s2):
    l1 = s1.split(" ")
    l2 = s2.split(" ")
    count = dict()
    for i in l1:
        count[i] = count.get(i,0) + 1
    for i in l2:
        count[i] = count.get(i,0) + 1
    return [word for word in count if count[word]==1]
```

```python
uncommon_words("hello bruh", "hello guys")
```

## Q6

```python
def duplicate_chars(s):
    count = {}
    for word in s.split(" "):
        count[word] = count.get(word, 0) + 1
    return [word for word in count if count[word]>1]
```

```python
duplicate_chars("this is a very good book but its very expensive and also this is very limited in stocks")
```

## Q7

```python
import string
def special_chars(s):
    for word in s:
        if word in string.punctuation:
            return True
    return False
```

```python
special_chars("hell*")
```
