##Q1

```python
C, H = 50, 30
D = input("Enter values of D : ")
D = D.split(",")
D = [int(x) for x in D]
Q = [((2*C*x)/H)**(1/2) for x in D]
for i in Q:
    print(i,sep=",")
```

##Q2

```python
X, Y = eval(input("Enter X and Y : "))
ans = [[0 for y in range(Y)] for x in range(X)]
for x in range(X):
    for y in range(Y):
        ans[x][y] = x*y
print(ans)
```

## Q5

```python
sent = input("Enter sentence : ")
sent = list(sent)
l, d = 0, 0
for i in sent:
    if i.isalpha():
        l = l+1
    if i.isdigit():
        d = d+1
    else:
        pass
print("letter : ",l,"  digits : ",d)
```

## Q6

```python
import re

passwords = input("Type in: ")
passwords = passwords.split(",")

accepted_pass = []
for i in passwords:
    if len(i) < 6 or len(i) > 12:
        continue
    elif not re.search("([a-z])+", i):
        continue
    elif not re.search("([A-Z])+", i):
        continue
    elif not re.search("([0-9])+", i):
        continue
    elif not re.search("([!@$%^&])+", i):
        continue
    else:
        accepted_pass.append(i)

print((" ").join(accepted_pass))
```
