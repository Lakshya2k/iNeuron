## Q1

```python
def equal(a, b, c):
    l = [a, b, c]
    d = dict()
    for i in l:
        d[i] = d.get(i,0) + 1
    return max(d.values())
```

```python
equal(3, 4, 3)
```

## Q3

```python
def mapping(l):
    d = dict()
    for i in l:
        d[i] = i.upper()
    return d 
```

```python
mapping(["a", "b", "c"])
```

## Q4

```python
def vow_replace(s, w):
    vowels = "aeiou"
    for v in vowels:
        s = s.replace(v, w)
    return s
```

```python
vow_replace("apples and bananas", "u")
```

## Q5

```python
def ascii_capitalize(s):
    ans = ""
    for i in s:
        if ord(i)%2 == 0:
            ans = ans + i.upper()
        else:
            ans = ans + i.lower()
    return ans
```

```python
ascii_capitalize("THE LITTLE MERMAID")
```
