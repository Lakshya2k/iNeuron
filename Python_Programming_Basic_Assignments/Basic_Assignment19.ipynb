## Q1

```python
def double_char(w):
    w1 = ""
    for i in w:
        w1 = w1 + i*2
    return w1
```

```python
double_char("1234!_")
```

## Q2

```python
def reverse(a):
    if type(a) == bool:
        return not(a)
    else:
        return "boolean expected"
```

```python
reverse(True)
```

```python
reverse(0)
```

## Q3

```python
def num_layers(n):
    ans = ((2**n)*0.5)/1000
    return str(ans) + "m"
```

```python
num_layers(21)
```

## Q4

```python
def index_of_caps(w):
    return [i for i in range(len(w)) if w[i].isupper()]
```

```python
index_of_caps("eQuINoX")
```

## Q5

```python
def find_even_nums(n):
    return [i for i in range(1, n+1) if i%2 == 0]
```

```python
find_even_nums(8)
```
