## Q1

```python
def find_sum(arr):
    sum = 0
    for i in arr:
        sum += i
    return sum
```

```python
find_sum([1,2,3,4,5,6,-6])
```

## Q2

```python
def find_max(arr):
    max = arr[0]
    for i in arr:
        if i>max:
            max=i
    return max
```

```python
find_max([-1,0,99,5,8])
```

## Q3

```python
def rotate_arr(arr, d):
    n = len(arr)
    temp = []
    i = 0
    while i < d:
        temp.append(arr[i])
        i = i + 1
    i = 0
    while d < n:
        arr[i] = arr[d]
        i = i + 1
        d = d + 1
    arr[:] = arr[: i] + temp
    return arr
```

```python
rotate_arr([11,22,33,44,55,66,77], 4)
```

## Q4

```python
def split_arr(arr, k):
    n = len(arr)
    for i in range(0, k):
        x = arr[0]
        for j in range(0, n-1):
            arr[j] = arr[j+1]
        arr[n-1] = x
    return arr
```

```python
split_arr([22,20,25,66,52,37], 3)
```
