## Q1

```python
def check_disarium(num):
    n = len(str(num))
    temp = num
    total = 0
    while temp>0:
        digit = temp%10
        total = total + digit**n
        n = n-1
        temp = temp//10
    if total == num:
        return "disarium"
    else:
        return "non-disarium"
```

```python
check_disarium(135)
```

```python
check_disarium(564)
```

## Q2

```python
def print_disarium(n):
    for i in range(1, n+1):
        if check_disarium(i) == "disarium":
            print(i, end=" ")
```

```python
print_disarium(100)
```

## Q3

```python
def square_sum(num):
    temp = num
    total = 0
    while temp>0:
        digit = temp%10
        total = total + digit**2
        temp = temp//10
    return total
def check_happy(num):
    slow = num
    fast = num
    while True:
        slow = square_sum(slow)
        fast = square_sum(square_sum(fast))
        if slow != fast:
            continue
        else:
            break
    return slow==1
```

```python
check_happy(13)
```

```python
check_happy(20)
```
