## Q1

```python
def divBy57(n):
    for i in range(n):
        if i%35 == 0:
            yield i
```

```python
n = int(input())
for i in divBy57(n):
    print(i)
```

## Q2

```python
n = int(input())
evens = (str(i) for i in range(n+1) if i%2==0)
print(",".join(list(evens)))
```

## Q3

```python
n = int(input())
ans = [0,1]
[ans.append(ans[-1]+ans[-2]) for n in range(n)]
ans = [str(i) for i in ans]
print(",".join(ans))
```

## Q5

```python
class Shape():
    def area(self):
        return 0
class Square(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b
    def area(self):
        return self.l*self.b
```

```python
sh, sq = Shape(), Square(8,8)
print(sh.area(),sq.area())
```
