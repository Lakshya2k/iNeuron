## Q1

```python
def extract_unique_values(d):
    d1 = {}
    for val in d.values():
        d1[val] = d1.get(val, 0) + 1
    return [val for val in d1 if d1[val]<2]
```

```python
extract_unique_values({"dog":"woof", "cat":"meow", "cow":"moo", "buffalo":"moo"})
```

## Q2

```python
def sum_items(d):
    total = 0
    for i in d.values():
        total = total + i
    return total
```

```python
sum_items({"a":100, "b":300, "c":200})
```

## Q3

```python
def merge_dict(d1, d2):
    return {**d1, **d2}
```

```python
merge_dict({"a":10, "b":20}, {"c":30, "d":40})
```

## Q6

```python
from collections import OrderedDict
def check_order(s, p):
    d = OrderedDict.fromkeys(s)
    plen = 0
    for key,val in d.items():
        if key==p[plen]:
            plen = plen + 1
        if plen==len(p):
            return True
    return False
```

```python
check_order("abcdacbdba", "dba")
```

## Q7

```python
def sort_dict(bykey, d):
    if bykey:
        return sorted(d.items(), key=lambda d:d[0])
    else:
        return sorted(d.items(), key=lambda d:d[1])
```

```python
sort_dict(True, {"a":3, "b":2, "c":1})
```

```python
sort_dict(False, {"a":3, "b":2, "c":1})
```
