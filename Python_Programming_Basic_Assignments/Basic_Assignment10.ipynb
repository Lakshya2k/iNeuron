## Q1

```python
def sum_list(l):
    total = 0
    for i in l:
        total = total + i
    return total
```

```python
sum_list([-1,-5,12,56,4,30])
```

## Q2

```python
def mul_list(l):
    result = 1
    for i in l:
        result = result * i
    return result
```

```python
mul_list([-1,-5,5,4])
```

## Q3

```python
def smallest_list(l):
    smallest = l[0]
    for i in l:
        if i<smallest:
            smallest = i
    return  smallest
```

```python
smallest_list([-9,0,45,6,3,-50])
```

## Q4

```python
def largest_list(l):
    largest = l[0]
    for i in l:
        if i>largest:
            largest = i
    return  largest
```

```python
largest_list([-50,12,54,99,100])
```

## Q7

```python
def print_even_list(l):
    for i in l:
        if i%2==0:
            print(i, end=" ")
```

```python
print_even_list([-2,1,5,6,12,31,50])
```

## Q8

```python
def print_odd_list(l):
    for i in l:
        if i%2!=0:
            print(i, end=" ")
```

```python
print_odd_list([-2,1,5,6,12,31,50])
```

## Q9

```python
def remove_empty_list(l):
    result = [i for i in l if i!=[]]
    return  result
```

```python
remove_empty_list([[],[],2,8,[]])
```

## Q10

```python
def copy_list(l):
    l1 = [i for i in l]
    return l1
```

```python
copy_list([1,2,3,4,5])
```

## Q11

```python
def count_list(l):
    d = {}
    for i in l:
        if i not in d:
            d[i] = 1
        else:
            d[i] = d[i] + 1
    return d
```

```python
count_list([1,1,2,3,4,4,4,5])
```
