## Q1

```python
def filter_list(l):
    return [x for x in l if type(x)!=str]
```

```python
filter_list([1, 2, "aasf", "1", "123", 123])
```

## Q2

```python
def reverser(s):
    s1 = ""
    for i in s:
        if i.islower():
            s1 = s1 + i.upper()
        if i.isupper():
            s1 = s1 + i.lower()
    return s1[::-1]    
```

```python
reverser("ReVeRsE")
```

## Q4

```python
def factorial(num):
    if num == (0 or 1):
        return 1
    else:
        return num*factorial(num-1)
```

```python
factorial(3)
```

## Q5

```python
def move_to_end(l, e):
    l1, l2 = [], []
    for i in l:
        if i == e:
            l2 = l2 +[i]
        else:
            l1 = l1 + [i]
    return l1 + l2
```

```python
move_to_end(["a", "a", "a", "b"], "a")
```
