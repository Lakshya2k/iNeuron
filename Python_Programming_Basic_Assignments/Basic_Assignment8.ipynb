## Q1

```python
def add_matrices(a, b):
    result = [[a[i][j] + b[i][j] for j in range(len(a[0]))] for i in range(len(a))]
    return result
```

```python
add_matrices([[12,7,3],[4,8,5],[11,10,1]],[[-1,4,7],[0,6,9],[1,1,1]])
```

## Q2

```python
def mul_matrices(a,b):
    result = [[sum(x*y for x,y in zip(i,j)) for j in zip(*b)] for i in a]
    return result
```

```python
mul_matrices([[12,7,3],[4,5,6],[7,8,9]],[[5,8,1,2],[6,7,3,0],[4,5,9,1]])
```

## Q3

```python
def transpose(a):
    result = [[a[j][i] for j in range(len(a))] for i in range(len(a[0]))]
    return result
```

```python
transpose([[1,2,3,4],[5,6,7,8]])
```

## Q4

```python
def sort_words(s):
    words = [word.lower() for word in s.split()]
    words.sort()
    return words
```

```python
sort_words("Hello everyone I can sort words alphabetically")
```

## Q5

```python
import string
def remove_punch(s):
    s1 = ""
    for char in s:
        if char not in string.punctuation:
            s1 = s1 + char
    return s1
```

```python
remove_punch("Hello!!! , how are u???")
```
