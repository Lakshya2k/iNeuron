## Q4

```python
import cmath #import complex module
def solve_quad(a, b, c):
    d = b**2 - 4*a*c
    r1, r2 = (-b-cmath.sqrt(d))/(2*a), (b-cmath.sqrt(d))/(2*a)
    return r1, r2
```

```python
solve_quad(1,2,4)
```
