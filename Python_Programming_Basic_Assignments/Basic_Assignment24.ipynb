## Q1

```python
def amplify(n):
    ans = []
    for i in range(1, n+1):
        if i%n == 0:
            ans = ans + [i*10]
        else:
            ans = ans + [i]
    return ans
```

```python
amplify(4)
```

## Q2

```python
def unique(l):
    d = dict()
    for i in l:
        d[i] = d.get(i, 0) + 1
    return [k for k,v in d.items() if v == 1][0]
```

```python
unique([3, 3, 3, 7, 3, 3])
```

## Q3

```python
import math
class Circle():
    def __init__(self, r):
        self.r = r
    def getArea(self):
        return math.pi*self.r**2
    def getPerimeter(self):
        return math.pi*self.r*2 
```

```python
c = Circle(4.44)
print(c.getArea(), c.getPerimeter())
```

## Q4

```python
def sort_by_length(l):
    return sorted(l, key = lambda x: len(x))
```

```python
sort_by_length(["Google", "Apple", "Microsoft"])
```

## Q5

```python
def is_triplet(a, b, c):
    l = [a, b, c]
    hi = max(l)
    lo = min(l)
    m = sum(l) - hi - lo
    return lo**2 + m**2 == hi**2
```

```python
is_triplet(3, 4, 5)
```
