## Q1

```python
def is_symmetrical(num):
    return str(num) == str(num)[::-1]
```

```python
is_symmetrical(11211)
```

```python
is_symmetrical(9939)
```

## Q2

```python
def multiply_nums(nums):
    nums = nums.split(", ")
    nums = [int(num) for num in nums]
    ans = 1
    for i in nums:
        ans = ans * i
    return ans
```

```python
multiply_nums("1, 2, 3, 4")
```

## Q3

```python
def square_digits(num):
    ans = ""
    for i in str(num):
        ans = ans + str(int(i)**2)
    return int(ans)
```

```python
square_digits(3212)
```

## Q4

```python
def setify(l):
    ans = []
    [ans.append(i) for i in l if i not in ans]
    return sorted(ans)
```

```python
setify([3, 3, 3, 2, 1])
```

## Q5

```python
def mean(num):
    count = 0
    total = 0
    for i in str(num):
        total = total + int(i)
        count = count + 1
    return total/count
```

```python
mean(666)
```
