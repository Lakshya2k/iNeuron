## Q1

```python
def divBy7(n):
    for i in range(n):
        if i%7==0:
            yield i
```

```python
for i in divBy7(100):
    print(i, sep="  ")
```

## Q2

```python
def getFreq(phrase):
    count = dict()
    for i in phrase.split(" "):
        count[i] = count.get(i, 0) + 1
    for k,v in sorted(count.items()):
        print(f"{k}:{v}")
```

```python
getFreq("New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3")
```

## Q3

```python
class Person():
    def getGender(self):
        return "Unknown"
class Male(Person):
    def getGender(self):
        return "Male"
class Female(Person):
    def getGender(self):
        return "Female"
```

```python
m, f = Male(), Female()
print(m.getGender(), f.getGender())
```

## Q6

```python
def binarySearch(arr, x, low, high):
    while low<high:
        mid = low + (high-low)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
binarySearch([3, 4, 5, 6, 7, 8, 9], 4, 0, 7)
```
