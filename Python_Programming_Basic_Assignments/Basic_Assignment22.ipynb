## Q1

```python
def list_operation(x, y, n):
    ans = []
    for i in range(x, y+1):
        if i%n == 0:
            ans = ans + [i]
    return ans
```

```python
list_operation(15, 20, 7)
```

```python
list_operation(1, 10, 3)
```

## Q2

```python
def simon_says(a, b):
    if a[:-1] == b[1:]:
        return True
    return False
```

```python
simon_says([1, 2], [5, 1])
```

```python
simon_says([1, 2, 3, 4, 5], [5, 5, 1, 2, 3])
```

## Q3

```python
def society_name(l):
    ans = ""
    for i in sorted(l):
        ans = ans + i[0]
    return ans
```

```python
society_name(["Phoebe", "Chandler", "Rachel", "Ross", "Monica", "Joey"])
```

## Q4

```python
def is_isogram(w):
    return len(w.lower()) == len(set(w.lower()))
```

```python
is_isogram("PaSsword")
```

```python
is_isogram("AlgoriSm")
```

## Q5

```python
def is_in_order(w):
    return w == "".join(sorted(w))
```

```python
is_in_order("edabit")
```

```python
is_in_order("xyzz")
```

```python
is_in_order("123")
```
