## Q1

```python
def filter_list(l):
    return [i for i in l if type(i) == int]
```

```python
filter_list([1, 2, 3, "a", "b", 4])
```

## Q2

```python
def add_indexes(l):
    return [i+e for i,e in enumerate(l)]
```

```python
add_indexes([1,2,3,4,5])
```

## Q3

```python
import math
def cone_volume(h, r):
    return math.pi*math.pow(r,2)*h*(1/3)
```

```python
cone_volume(15, 6)
```
