```python
def stutter(word):
    f2 = word[:2]
    return f"{f2}... {f2}... {word}?"
```

```python
stutter("outstanding")
```

## Q2

```python
import math
def rad2deg(angle):
    return round(angle*(180/math.pi), 1)
```

```python
rad2deg(1)
```

## Q3

```python
def is_curzon(num):
    return bool(not(1+2**num)%(1+2*num))
```

```python
is_curzon(14)
```

```python
def area_of_hexagon(s):
    return (3*(3**(1/2))*(s**2))/2
```

```python
area_of_hexagon(3)
```
